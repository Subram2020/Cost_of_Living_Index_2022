```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Cost_of_Living_Index_2022.csv")
```

```python
#Basic Statistical details of Cost of Living 2022
df.describe()
```

```python
df.head(10)
```

```python
#Top 10 Cost of Living Plus Rent Index 
df.nlargest(10,['Cost of Living Plus Rent Index'])
```

```python
#Top 10 Lowest Cost of Living Plus Rent Index
df.nsmallest(10,['Cost of Living Plus Rent Index'])
```

```python
#Top 10 lowest countries with least local purchasing Power
df.nsmallest(10,['Local Purchasing Power Index'])
```

```python
#Top 10 expensive countries to dine in restaurants
df.nlargest(10,['Restaurant Price Index'])
```

```python
#Distribution of Groceries Index
plt.figure(figsize=(20,4))
sns.displot(data=df,x='Groceries Index',color = 'Black', kde=True, bins = 20,edgecolor = 'Yellow', lw = 4)
```

```python
#Relationship between Cost of Living & Rent
plt.figure(figsize=(12,8))
sns.scatterplot(x='Rent Index',y='Cost of Living Index',data=df,color = 'Red')
```

```python
high_cost_living = df.nlargest(n=10,columns=['Cost of Living Index'])
high_cost_living
```

```python
plt.figure(figsize=(15,8))
plt.plot('Country','Cost of Living Index',data = high_cost_living,color='b')
plt.title('Top countries by higest cost of living index')
plt.show()
```

```python
low_cost_living = df.nsmallest(n=10,columns=['Cost of Living Index'])
low_cost_living
```

```python
plt.figure(figsize=(18,8))
plt.plot('Country','Cost of Living Index',data = low_cost_living,color='g',ls = '-', marker='s',markersize=20,markerfacecolor="yellow",markeredgecolor="brown")
plt.title('Top countries by lowest cost of living index')
plt.show()
```

```python
high_rent = df.nlargest(n=10,columns=['Rent Index'])
high_rent
```

```python
plt.figure(figsize=(15,8))
plt.plot('Country','Rent Index',data = high_rent,color = 'blue',marker='+',markersize=20,markerfacecolor = 'green',markeredgecolor="brown")
plt.title('Top expensive rent countries')
plt.show()

```

```python
#Relaltionship between Restaurant Price & Local Purchasing Power
plt.figure(figsize=(12,6))
sns.jointplot(x='Restaurant Price Index',y='Local Purchasing Power Index',data=df, kind = 'hex')
```

```python
#Correlation of Cost of Living Index 2022
sns.heatmap(df.drop('Country',axis = 1).corr(), cmap = "viridis", annot = True)
```
